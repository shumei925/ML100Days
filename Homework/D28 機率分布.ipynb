### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
from scipy.stats import binom
import numpy as np

# p(x=正面)=0.5
# n=100表示丟100次
# x為可能的結果，也就是0,1,2...,50個正面

n, p=100,0.5
x=np.arange(0,101)
w=binom.pmf(x,n,p)

# 出現50次正面的機率
print(w[50:51])
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1)

n, p = 100, 0.5
x = np.arange(0,101)

ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)

ax.legend(loc='best', frameon=False)
plt.show()
```

### Discrete uniform distribution 離散型分布
p(x) = 1/n

```python
from scipy.stats import randint
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1)

# np.arange: return evenly spaced values
low,high = 1, 7
x = np.arange(low,high)
print('樣本點',x)

# pmf probability mass function: 1/(high-low)
# randint.pmf(樣本數, 最小樣本值, 最高樣本值+1)
y=randint.pmf(x, low, high)
print('pmf',y)
ax.plot(x,y, 'bo', ms=8, label='randint pmf') 
ax.vlines(x, 0, y, colors='b', lw=5, alpha=0.5)
ax.legend(loc='best', frameon=False)

plt.show()
```

```python
# cdf Cumulative distribution function
prob = randint.cdf(x, low, high)
print('累積機率',prob)

# 隨機產生樣本
r = randint.rvs(low, high, size=5)
print('在指定範圍內隨機取5個數字:',r)

# ppf percent point function: 分位數
print('1-6的分位數Q2:',randint.ppf(0.5, low, high))
```

### Bernoulli distribution 伯努利分布
當一個試驗有兩種可能性，兩機率為固定

p(x=0)=p ；p(x=1)=1-p ；做n次試驗

當 n = 1 時是伯努利分布 ；當 n > 1 時是二項式分布

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1)

# p(x=1)=0.3
p = 0.3
x = np.arange(0,2)
print('x',x)
ax.plot(x, bernoulli.pmf(x, p), 'bo', ms=8, label='bernoulli pmf')
ax.vlines(x, 0, bernoulli.pmf(x, p), colors='b', lw=5, alpha=0.5)
ax.legend(loc='best', frameon=False)

plt.show()
```

### binomial distribution 二項式分布
當一個試驗有兩種可能性，兩機率為固定

p(x=0)=p ；p(x=1)=1-p ；做n次試驗

當 n = 1 時是伯努利分布 ；當 n > 1 時是二項式分布

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1)

n, p = 6, 0.2

x = np.arange(0,7)
print('p(x=1)=0.2, p(x=0)=0.8')
print('做6次可能的結果:',x)
print('每個結果的機率:',binom.pmf(x, n, p))

ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)

ax.legend(loc='best', frameon=False)
plt.show()
```
